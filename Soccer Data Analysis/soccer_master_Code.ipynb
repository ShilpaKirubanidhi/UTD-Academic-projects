```python
#Creating master dataframe and adding necessary fields

import pandas as pd
import datetime
import matplotlib.pyplot as plt
import numpy as np

csv_0 = pd.read_csv('E0.csv')
csv_1 =pd.read_csv('E0 (1).csv')
csv_2 =pd.read_csv('E0 (2).csv')
csv_3 =pd.read_csv('E0 (3).csv')
csv_4 =pd.read_csv('E0 (4).csv')
csv_5 =pd.read_csv('E0 (5).csv')
csv_6 =pd.read_csv('E0 (6).csv')
csv_7 =pd.read_csv('E0 (7).csv')
csv_8 =pd.read_csv('E0 (8).csv')
csv_9 =pd.read_csv('E0 (9).csv')
csv_10 =pd.read_csv('E0 (10).csv')
csv_11 =pd.read_csv('E0 (11).csv')
csv_12 =pd.read_csv('E0 (12).csv')
csv_13 =pd.read_csv('E0 (13).csv')
csv_14 =pd.read_csv('E0 (14).csv')
csv_15 =pd.read_csv('E0 (15).csv')
csv_16 =pd.read_csv('E0 (16).csv')
csv_17 =pd.read_csv('E0 (17).csv')
csv_18 =pd.read_csv('E0 (18).csv',encoding= 'unicode_escape')
csv_19 =pd.read_csv('E0 (19).csv')
csv_20 =pd.read_csv('E0 (20).csv')
csv_21 =pd.read_csv('E0 (21).csv')
csv_22 =pd.read_csv('E0 (22).csv')
csv_23 =pd.read_csv('E0 (23).csv')
csv_24 =pd.read_csv('E0 (24).csv')
csv_25 =pd.read_csv('E0 (25).csv')
csv_26 =pd.read_csv('E0 (26).csv')
csv_27 =pd.read_csv('E0 (27).csv')
csv_28 =pd.read_csv('E0 (28).csv')
csv_29 =pd.read_csv('E0 (29).csv')


filter_cols=['Date','HomeTeam','AwayTeam','FTHG','FTAG','FTR']

df_0 =csv_0[filter_cols].dropna(how="all")
df_1 =csv_1[filter_cols].dropna(how="all")
df_2 =csv_2[filter_cols].dropna(how="all")
df_3 =csv_3[filter_cols].dropna(how="all")
df_4 =csv_4[filter_cols].dropna(how="all")
df_5 =csv_5[filter_cols].dropna(how="all")
df_6 =csv_6[filter_cols].dropna(how="all")
df_7 =csv_7[filter_cols].dropna(how="all")
df_8 =csv_8[filter_cols].dropna(how="all")
df_9 =csv_9[filter_cols].dropna(how="all")
df_10 =csv_10[filter_cols].dropna(how="all")
df_11 =csv_11[filter_cols].dropna(how="all")
df_12 =csv_12[filter_cols].dropna(how="all")
df_13 =csv_13[filter_cols].dropna(how="all")
df_14 =csv_14[filter_cols].dropna(how="all")
df_15 =csv_15[filter_cols].dropna(how="all")
df_16 =csv_16[filter_cols].dropna(how="all")
df_17 =csv_17[filter_cols].dropna(how="all")
df_18 =csv_18[filter_cols].dropna(how="all")
df_19 =csv_19[filter_cols].dropna(how="all")
df_20 =csv_20[filter_cols].dropna(how="all")
df_21 =csv_21[filter_cols].dropna(how="all")
df_22 =csv_22[filter_cols].dropna(how="all")
df_23 =csv_23[filter_cols].dropna(how="all")
df_24 =csv_24[filter_cols].dropna(how="all")
df_25 =csv_25[filter_cols].dropna(how="all")
df_26 =csv_26[filter_cols].dropna(how="all")
df_27 =csv_27[filter_cols].dropna(how="all")
df_28 =csv_28[filter_cols].dropna(how="all")
df_29 =csv_29[filter_cols].dropna(how="all")


df_list=[df_0,
df_1,
df_2,
df_3,
df_4,
df_5,
df_6,
df_7,
df_8,
df_9,
df_10,
df_11,
df_12,
df_13,
df_14,
df_15,
df_16,
df_17,
df_18,
df_19,
df_20,
df_21,
df_22,
df_23,
df_24,
df_25,
df_26,
df_27,
df_28,
df_29
]

for df in df_list:
    df['Year'] = pd.DatetimeIndex(df['Date'], dayfirst=True).year
    df['Season']=str(df['Year'].min())+"/"+str(df['Year'].max())
    df['Date']=pd.to_datetime(df['Date'],dayfirst=True)
    df['Week_Of_Year'] = df['Date'].dt.isocalendar().week
    
master_df=pd.concat([df_0,
df_1,
df_2,
df_3,
df_4,
df_5,
df_6,
df_7,
df_8,
df_9,
df_10,
df_11,
df_12,
df_13,
df_14,
df_15,
df_16,
df_17,
df_18,
df_19,
df_20,
df_21,
df_22,
df_23,
df_24,
df_25,
df_26,
df_27,
df_28,
df_29],ignore_index=True)



master_df.loc[master_df['FTR']=='A','Winner']=master_df['AwayTeam']
master_df.loc[master_df['FTR']=='H','Winner']=master_df['HomeTeam']
master_df.loc[master_df['FTR']=='D','Winner']='Draw'

master_df.loc[master_df['FTR']=='A','Winning_Team_Goals']=master_df['FTAG']
master_df.loc[master_df['FTR']=='H','Winning_Team_Goals']=master_df['FTHG']
master_df.loc[master_df['FTR']=='D','Winning_Team_Goals']=master_df['FTHG']

master_df.loc[master_df['FTR']=='A','Loser']=master_df['HomeTeam']
master_df.loc[master_df['FTR']=='H','Loser']=master_df['AwayTeam']
master_df.loc[master_df['FTR']=='D','Loser']='Draw'

master_df.loc[master_df['FTR']=='A','Losing_Team_Goals']=master_df['FTHG']
master_df.loc[master_df['FTR']=='H','Losing_Team_Goals']=master_df['FTAG']
master_df.loc[master_df['FTR']=='D','Losing_Team_Goals']=master_df['FTHG']

master_df['Total_FT_Goals']=master_df.FTHG+master_df.FTAG
```

```python
#Q1 - best performing team (Most #of victories)
#Q2 - worst performing team (Least #of victories)
#nodraws_df is data without draw matches
def victories_stats(master_df):
    nodraws_df=master_df[master_df['FTR']!='D']
    print("Based on number of victories:")
    print(nodraws_df['Winner'].value_counts())
    print("The best performing team is:",nodraws_df['Winner'].value_counts().idxmax())
    print("The worst performing team is:",nodraws_df['Winner'].value_counts().idxmin())

victories_stats(master_df)
print("---------------------------------------------------")
```

```python
#Q3 - Which team scored the most goals in each season
#grouping data by season, merging and finding the team with maximum goals
df_home=pd.DataFrame(master_df.groupby(['Season','HomeTeam'])['FTHG'].sum().reset_index())
df_away=pd.DataFrame(master_df.groupby(['Season','AwayTeam'])['FTAG'].sum().reset_index())
df_home.rename(columns={'HomeTeam': 'Team', 'FTHG': 'Total_Goals'}, inplace=True)
df_away.rename(columns={'AwayTeam': 'Team', 'FTAG': 'Total_Goals'}, inplace=True)
new_df=pd.concat([df_home,df_away],ignore_index=True)
df_team_goals=pd.DataFrame(new_df.groupby(['Season','Team'])['Total_Goals'].sum().reset_index())
maxval=pd.DataFrame(df_team_goals.groupby(['Season'])['Total_Goals'].max().reset_index())
print("Team with maximum number of goals in each season is:")
df_merged_max=pd.merge(maxval,df_team_goals,how='inner',on=('Season','Total_Goals'))
print(df_merged_max)

print("---------------------------------------------------")
```

```python
#Q4 - Which team scored the least goals overall
#teams and the total goals (overall stats)

def overall_stats(master_df):
    df_home_o=pd.DataFrame(master_df.groupby('HomeTeam')['FTHG'].sum().reset_index())
    df_away_o=pd.DataFrame(master_df.groupby('AwayTeam')['FTAG'].sum().reset_index())

    df_home_o.rename(columns={'HomeTeam': 'Team', 'FTHG': 'Total_Goals'}, inplace=True)
    df_away_o.rename(columns={'AwayTeam': 'Team', 'FTAG': 'Total_Goals'}, inplace=True)

    new_df_o=pd.concat([df_home_o,df_away_o],ignore_index=True)

    df_team_goals_o=pd.DataFrame(new_df_o.groupby('Team')['Total_Goals'].sum().reset_index())

    
    maxval=df_team_goals_o['Total_Goals'].max(axis=0)

    minval=df_team_goals_o['Total_Goals'].min(axis=0)

    #print("Team with maximum number of goals overall is:")
    #print(df_team_goals_o.loc[df_team_goals_o['Total_Goals']==maxval])

    print("Team with minimum number of goals overall is:")
    print(df_team_goals_o.loc[df_team_goals_o['Total_Goals']==minval])


overall_stats(master_df)

print("---------------------------------------------------")
```

```python
#Q5 - Which is the best performing team in the last 3 years? Is it the same team that scored most goals?

latest3_df=master_df[master_df['Year']>=2020]
print("For last 3 years Jan 2020 to Feb 2023")
victories_stats(latest3_df)
print("Man city is the best performing team, which is the same as the team that scored most goals (from Question 3).")
print("---------------------------------------------------")
```

```python
#Q7 Plot the distribution of goals per match for every year (winner vs loser)

master_df1=pd.DataFrame(master_df,columns=['Year','Winning_Team_Goals','Losing_Team_Goals'])
pltmaster_df1=pd.DataFrame(master_df1.groupby('Year')['Winning_Team_Goals'].sum().reset_index())
pltmaster_df2=pd.DataFrame(master_df1.groupby('Year')['Losing_Team_Goals'].sum().reset_index())
fig = plt.figure(figsize = (16, 5))
plt.bar( pltmaster_df1.Year-0.2, pltmaster_df1.Winning_Team_Goals, width = 0.4)
plt.bar( pltmaster_df1.Year+0.2, pltmaster_df2.Losing_Team_Goals, width = 0.4)
plt.title("Distribution of goals for every year Winner Vs. Loser")
plt.xlabel("Year")
plt.ylabel("Total Goals")
plt.legend(["Winning_Team_Goals", "Losing_Team_Goals"],bbox_to_anchor =(1.15, 1.15),loc="upper right")
plt.show()

print("---------------------------------------------------")
```
